## **Trabalhando com planilhas de Excel** *italicized text*

```python
import pandas as pd
```

```python
#Leitura dos 5 arquivos de forma permanente no drive
df1 = pd.read_excel("/content/drive/MyDrive/Curso_Python_Pandas_Digital_Innovation-master/datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Curso_Python_Pandas_Digital_Innovation-master/datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/Curso_Python_Pandas_Digital_Innovation-master/datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Curso_Python_Pandas_Digital_Innovation-master/datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Curso_Python_Pandas_Digital_Innovation-master/datasets/Salvador.xlsx")
```

```python
#Juntando todos os arquivos através da concatenação
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Alterando o tipo de um dado da coluna - para float64
df["LojaID"] = df["LojaID"].astype("float64")
df.dtypes
```

```python
#Alteranod o tipo de um dado da coluna - para object
df["LojaID"] = df["LojaID"].astype("object")
df.dtypes
```

## **Tratando de valores ausentes**

```python
#Subistituindo os valores nulos pela média da coluna de "Vendas"
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
#Verificando a média de vendas
df["Vendas"].mean()
```

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
df.sample(15)
```

```python
#Subistituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
df.isnull().sum()
```

```python
#Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos com base em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

## **Criando colunas novas**

```python
#Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
#Caso na planilha não houvesse a coluna de quantidade, mas sim de receita, para identificar utilizariámos o seguinte
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#retornando o top 3 com base na coluna Receita
df.nlargest(3, "Receita")
```

```python
#Retornando as 3 piores de Receita
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

## **Trabalhando com Datas**

```python
#Transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
df.dtypes
```

```python
#Transformando a coluna de data (inteiro) em formato de datetime
df["Data"] = pd.to_datetime(df["Data"])
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_venda"] = df["Data"].dt.year
df.sample(5)
```

```python
#Extraindo o dia e o mês da venda
df["dia_venda"], df["mes_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
df.sample(10)
```

```python
#Retornando a data mais antiga do conjunto de dados
df["Data"].min()
```

```python
#Retornando a data mais recente do conjunto de dados
df["Data"].max()
```

```python
#Calculando a diferença de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month ==3)]
```
