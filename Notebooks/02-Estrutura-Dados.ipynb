```python
#Criando uma nova lista de valores
lista = [500, 700, 800, 30, 70, 60, 580]
```

```python
#Imprimindo na tela o valor máximo da lista
max(lista)
```

```python
#Imprimindo na tela o valor mínimo da lista
min(lista)
```

```python
#Tuplas = São imutáveis, elas não permitem alterar valor dentro de uma tupla. Elas usam parênteses como sintaxe.
tp = ("Banana", "Maçã", 10, 50) 
```

```python
#Retornando o segundo elemento
tp[2]
```

```python
#Determinando o valor de determinado elemento de uma tupla
tp.count("Maçã")
```

```python
#Listando uma tupla por intervalo
tp[0:3]
```

```python
#Para criar um dicionário utilizamos as {}
dc = {"Maçã":20, "Banana":10, "Laranja":15, "Uva":5} #Os dicionários trabalham com um conceito chave valor.
dc
```

```python
dc["Maçã"]
```

```python
#Atualizando o valor da Maçã e Imprimindo a lista
dc["Maçã"] = 75
dc
```

```python
#Retornando todas as chaves do dicionário
dc.keys()
```

```python
#Retornando os valores do dicionário
dc.values()
```

```python
#Verificando se já existe uma chave no dicionário e caso não exista inserir
dc.setdefault("Manga", 90)
dc
```
